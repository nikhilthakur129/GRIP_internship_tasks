```python
#Task 1 Student Marks Predictions Using Simple Linear Regression
#Here we installed several libraries for the regression
```

```python
from sklearn import linear_model
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#loading the data in the variable named as s_data
```

```python
#Getting the structure of the data for further analysis
```

```python
s_data = pd.read_csv("student_scores.csv")
print("data imported successfullly")
s_data.head(20)
```

```python
#Plotting the graph of the given data understandig the distribution of the data
```

```python
plt.xlabel("No of hours per day")
plt.ylabel("Marks scored")
plt.title("Hours Vs Marks Graph")
plt.scatter(x="Hours",y="Scores",data=s_data,marker=".",color= "black")
plt.show()
```

```python
#creating features and labels
```

```python
x = s_data.iloc[:,:-1].values
y = s_data.iloc[:,1].values
```

```python
#splitting the datasets in train and test 
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.2,random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
LR = LinearRegression()
```

```python
#feeding the dataset to the LR(instance)
```

```python
LR.fit(x_train,y_train)
```

```python
#prediction 

print(x_test)
y_pred = LR.predict(x_test)
```

```python
#task to predict marks scored if student studeid for 9.25 hours

input_Hours = np.array([[9.25],[0]])
input_Hours.reshape(-1,1)
result = LR.predict(input_Hours)
print("student spend  per day,they got {1} marks".format(input_Hours,result[0]))
```

```python
#Evalution of the model
```

```python
from sklearn import metrics
print("The mean squared error ",metrics.mean_squared_error(y_test,y_pred))
```
